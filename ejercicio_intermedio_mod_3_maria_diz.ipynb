```python
# Necesitamos un diccionario donde las Keys son los nombres antiguos y los values son los nombres nuevos.
# No podemos construir un diccionario a mano, tenemos que hacer un *dict comprehension*
# vamos a iterar por todas las columnas del DataFrame y a cada una de ellas la pondremos en minúsula
# usaremos el metodo rename para aplicar los nuevos nombres

columnas_mod = {columna:columna.lower() for columna in df.columns}

columnas_mod
```

```python
# una vez que tenemos el diccionario vamos a aplicar el método rename

df.rename(columns = columnas_mod, inplace = True)

# comprobamos que se han cambiado los nombres de las columnas
df.head(2)
```


- Haciendo el análisis exploratorio te deberías haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Crea una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
# creamos la funcion para cambiar comas por puntos, estas variables son tipo object por que pasaran a ser float
# creo un lista con las columnas a modificar

def cambiar_dato(dato):
    try:
        return float(dato.replace(",","."))
    except:
        return np.nan
```

```python
lista_columnas = ['danceability','energy','key','loudness','speechiness','acousticness' ,'instrumentalness']

for columna in lista_columnas:
    df[columna] = df[columna].apply(cambiar_dato)
```

- Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haz lo mismo para las columnas de "title" y "album".

```python
# visualizamos una muestra aleatoria de las columnas que nos afecta para el cambio

df[["title", "album", "track"]].sample(4)
```

```python
# creamos la funcion para cambiar a minusculas

def cambiar_minusculas(contenido):
    return contenido.str.lower()

# como son varias columnas creamos un bucle for para iterar por ellas

lista_columnas = ["title", "album", "track"]

for columna in lista_columnas:
    df[columna] = cambiar_minusculas(df[columna])

# visualizamos una muestra para ver el cambio
df[["title", "album", "track"]].sample(2)
```

- Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los videos de música de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, querremos contestar a la siguiente pregunta: ¿Son aquellos artistas que tienen más visitas los que más likes tienen? Para solucionar este ejercicio deberéis:

    - Agrupa por artista y calcular la media de visitas y de likes.
    - Quedate con los 10 artistas que más visitas han tenido.
    - Quedate con los 10 artistas que más likes han tenido.
    - BONUS Haced lo mismo para los datos de tipo "single".

```python
mas_visitas = df_album.groupby('artist')[['views','likes']].mean().sort_values(by ='views', ascending = False)
mas_likes = df_album.groupby('artist')[['views','likes']].mean().sort_values(by ='likes', ascending = False)
mas_visitas.head(10)
```

```python
mas_likes.head(10)
```

- BONUS Haced lo mismo para los datos de tipo "single".

```python
mas_visitas_single = df_single.groupby('artist')[['views','likes']].mean().sort_values(by ='views', ascending = False)
mas_likes_single = df_single.groupby('artist')[['views','likes']].mean().sort_values(by ='likes', ascending = False)
mas_visitas_single.head(10)
```
